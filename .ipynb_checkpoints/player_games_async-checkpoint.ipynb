```python
def create_coroutines_for_player_games(usernames, from_date, to_date):
    # If from_date or to_date are not datetime objects, raise an error and stop the function.
    if not isinstance(from_date, datetime) or not isinstance(to_date, datetime):
        raise TypeError("Both from_date and to_date must be datetime objects")

    # Force from_date and to_date to have the first of any given month
    from_date = datetime(from_date.year, from_date.month, 1)
    to_date = datetime(to_date.year, to_date.month, 1)

    # If from_date is after to_date, raise an error and stop the function.
    if from_date > to_date:
        raise ValueError("from_date must be before to_date")
 
    # Create a iteratable date from the from_date, that will count towards to date.
    iter_date = from_date

    # Create a iter_list of the dates with from_date -> to_date.
    iter_list = []
    while iter_date <= to_date:
        # Add the current iter_date into the iter_list
        iter_list.append(iter_date)

        # Move the iter_date to the next month
        new_month = iter_date.month + 1
        new_year = iter_date.year

        if new_month > 12:
            new_month = 1
            new_year += 1
        iter_date = datetime(new_year, new_month, 1)

    # For each username in usernames
    all_cors = []
    for username in usernames:
        cors = [get_player_games_by_month(username, N.year, N.month) for N in iter_list]   
        all_cors = all_cors + cors

    return all_cors
```

```python
from_date = datetime(2022, 1, 1)
to_date = datetime(2024, 1, 1)

cors = create_coroutines_for_player_games([usernames[0],usernames[1], usernames[2],usernames[3]], from_date, to_date)
```

```python
async def gather_coroutines(*coroutines, placeholder_value=None, conc_limit=10):
    semaphore = asyncio.Semaphore(conc_limit)

    total_coroutines = len(coroutines)
    completed_coroutines = 0

    async def run_sem_cor(cor):
        nonlocal completed_coroutines
        async with semaphore:
            try:
                result = await cor
            except Exception as e:
                print(f"Error occurred: {e}")
                result = placeholder_value
            completed_coroutines += 1
            if completed_coroutines % 100 == 0:
                print(f"Completed {completed_coroutines}/{total_coroutines} coroutines")
            return result

    return await asyncio.gather(*(run_sem_cor(c) for c in coroutines))
```

```python
values = await gather_coroutines(*cors, conc_limit=100)
```

```python
# https://stackoverflow.com/questions/47675410/python-asyncio-aiohttp-valueerror-too-many-file-descriptors-in-select-on-win
# https://stackoverflow.com/questions/48483348/how-to-limit-concurrency-with-python-asyncio/61478547#61478547

async def gather_coroutines(coroutines, placeholder_value=None, conc_limit=10):
    semaphore = asyncio.Semaphore(conc_limit)

    total_coroutines = len(coroutines)
    completed_coroutines = 0

    async def run_sem_cor(cor):
        nonlocal completed_coroutines
        async with semaphore:
            # return await cor
            result = await cor
            completed_coroutines += 1
            if completed_coroutines % 100 == 0:
                print(f"Completed {completed_coroutines}/{total_coroutines} coroutines")
            return result
    
    tasks = [run_sem_cor(cor) for cor in coroutines]
    print(len(tasks))
    done, _ = await asyncio.wait(tasks, return_when=asyncio.ALL_COMPLETED)

    responses = []
    for task in done:
        try:
            response = task.result()
        except ChessDotComError as e:
            print(f"Error occurred: {e}")
            response = placeholder_value
        responses.append(response)
        
    return responses
    
```
